```python
import json
import matplotlib.pyplot as plt
```

```python
with open('tournament_data.json', 'r') as f:
    tournament_data = json.load(f)

memorials = [t for t in tournament_data[::-1] if "memorial" in t[2].lower()]

players = [int(t[7]) for t in memorials]
rounds = [int(t[6]) for t in memorials]

years = list(map(str, range(2005, 2005 + len(players))))
```

```python
plt.figure(figsize=(10, 4))
plt.bar(years, players, label="Players")
plt.plot(years, [rs * 4 for rs in rounds], color="red", label="Rounds")
plt.xlabel('Years')
plt.xticks(rotation=90)
plt.ylabel('Players')
plt.show()
```
